# 1. Seawater contains charged ions
As we read in Chapter 7, seawater contains about 35 grams (g) of **inorganic ions** per kilogram (kg) of seawater. (As a reminder, 1 kg = 1000 g). An **ion** is a charged particle, typically containing one or a few atomic nuclei. For example, the most abundant ion in seawater, **chloride (Cl<sup>-</sup>)**, is identical to a chlorine (Cl) atom plus an exta, negatively chaged, electron (0 + (-1) = -1). The second most abundant ion in seawater is the **sodium ion (Na<sup>+</sup>)**, which is identical to the sodium atom minus one electron (0 - (-1) = +1). Under certain conditions, these inorganic ions can come together to form neutral **minerals** known as **"salts"**. For example, sodium and chloride ions can come together to form the neutral compound **sodium chloride** (+1 + -1 = 0). If formed under natural conditions, the compound **sodium chloride** is known as the mineral, **halite**. Importantly, the word **"salt"** only refers to a certain type of solid compund. Once these solid compounds are dissolved in water, either in the ocean or on your tongue, they break down into charged ions. Thus, while seawater is salty and has a measureable property called **"salinity"**, it does not contain **"salt"** *per se*. Also important is the fact that neutral minerals (like halite) likewise contain charged ions, not neutral atoms. Sodium ions in halite are still short one electron relative to neutral sodium atoms. Conversely, chloride ions in halite still have one extra electron relative to neutral chlorine atoms. The compound, however, is neutral (+1 + -1 = 0)  

It is important to note that some ions contain more than one atomic nucleus. Accordingly, these are called **polyatomic ions**. Important examples of **polyatomic ions** in seawater include **sulfate (SO<sub>4</sub><sup>2-</sup>)**, **carbonate (CO<sub>3</sub><sup>2-</sup>)**, **nitrate (NO<sub>3</sub><sup>-</sup>)**, and **phosphate (PO<sub>4</sub><sup>3-</sup>)**.   

In this problem set, we will also pay close attention to the behavior of the ions, **carbonate (CO<sub>3</sub><sup>2-</sup>)** and **bicarbonate (HCO<sub>3</sub><sup>-</sup>)**, both with respect to each other and with respect to neutral **carbon dioxide (CO<sub>2</sub>)** gas. The carbonate ion contains one carbon atom and three oxygen atoms plus two extra electrons relative to the neutrally charged atoms. A bicarbonate ion is like a carbonate ion plus a positively chaged **hydrogen ion (H<sup>+</sup>)** (-2 + +1 = -1). The concentration and chemical behavior of the hydrogen ion (H<sup>+</sup>) is particularly imporant in determining the acid/base behavior of an aqueous solution (*i.e.* the **pH**) . 


## 1.1. Minerals can precipitate from ions in seawater 
If you evaporate seawater, different salts will form (precipitate) at different points in the evaporation process. Exactly which minerals form, how much of each, and in what order depends on the composition of seawater and temperatures and pressures under which evaporation takes place. If water is evaporated from seawater at temperatures and pressures similar to those encountered on the Earth's surface, the first mineral to precipitate is typically: 
* **Calcite**, a calcium carbonate mineral (CaCO<sub>3</sub>)  

Seawater is saturated with respect to calcite, so calcite precipitates as soon as you start evaporating water, and can even precipitate from unevaporated seawater. Many marine organisms use **calcite** and its close cousin **aragonite** (another calcium carbonate mineral, also CaCO<sub>3</sub>) to make shells and other useful body parts. Your bones and teeth also contain significant amounts of calcite.  

After about 80% of the water has been evaporated, the remaining brine will precipitate:  
* **Gypsum**, a hydrous calcium sulfate mineral (CaSO<sub>4</sub>·2H<sub>2</sub>O)  

Gypsum is very useful for making plaster and chalk.

After about 90% of the water has been evaporated, the remaining brine will precipitate:

* **Halite**, a sodium chloride mineral (NaCl)  

Halite is composed of sodium (Na) and chloride (Cl), the two most abundant ions dissolved in seawater. Halite is also the mineral we typically think of when we talk about "salt". Unlike marine organisms, many land animals need to supplement their diets with sodium and chloride, which may explain humans' attraction to salty flavors. However, humans have gotten so good at finding, making, and trading **halite** that most humans should probably consume less of this mineral (technically the Na<sup>+</sup> and Cl<sup>-</sup> ions), not more. 

Continued evaporation of seawater will lead to the precipitation of a variety of salts, primarily sulfates, that are collectively known as **"bitterns"**. Some examples of these **"bitterns"** include:  

* **Glauberite**, a sodium calcium sulfate mineral (Na<sub>2</sub>Ca(SO<sub>4</sub>)<sub>2</sub>)
* **polyhalite**, a hydrous potassium calcium magnesium sulfate (K<sub>2</sub>Ca<sub>2</sub>Mg(SO<sub>4</sub>)<sub>4</sub>·2H<sub>2</sub>O)
* **Anhydrite**, an anhydrous calcium sulfate mineral (CaSO<sub>4</sub>)
* **Epsonite**, a hydrous magnesium sulfate mineral (MgSO<sub>4</sub> · 7H<sub>2</sub>O)
* **Kieserite**, another hydrous magnesium sulfate mineral (MgSO<sub>4</sub> 4· H<sub>2</sub>O)
* **Carnallite**, a potassium magnesium chloride mineral (KMgCl<sub>3</sub>)
* **Bischofite**, a hydrous magnesium chloride mineral (MgCl<sub>2</sub>·6H<sub>2</sub>O)  

As their name suggests, these minerals collectively taste bitter and are therefore undesireable in salt production. Human processes to collect "salt", whether artesianal or industrial or working from current seawater or geologically preserved brines or salt deposits, serve to control the mining and evaporation process and thereby separate **halite** from **calcite** and **gypsum** on the one hand and **"bitterns"** on the other. "Sea salt" that you might buy at the store primarily contains halite, but would tend not to contain much calcite, gypsum or bitterns. Thus, seawater cannot be made by simply adding water to "sea salt". # 3. Mass vs. Number of Molecules
The table in Chapter 7 is written in g/kg. In other words, units of mass. As you can see by both the tables and the pie charts, the dominant element by mass is oxygen. However, the **chemical formula for water (H<sub>2</sub>O)** tells us that there are actually two **hydrogen atoms (H)** for every one **oxygen atom (O)**. Why the discrepancy? Well, clearly one oxygen atom has a greater mass than one hydrogen atom. On average, oxygen atoms have a mass of 15.999 g/mol while hydrogen atoms, on average, have a mass of 1.008 g/mol. Another name for the unit **g/mol** is the **"atomic mass unit"** (abbreviated 'amu' or  'u'). 

## **But wait!!!**

What is this **"mol"** thing?

## 3.1. Moles and atomic mass
A **"mole"** is a number, just like a "dozen" is a number (12) or a "hundred" is a number (100). Specifically, a mole (abbreviated "mol") is **Avegadro's Number**, which is **6.02214076×10<sup>23</sup>**. Recalling **scientific notation**, you can see right away that this is a huge number, something like **six hundred billion trillion**. Just like you could have "five dozen" cookies, or "five hundred" cookies, you could also have "five moles" of cookies. Adopting an 11 g sandwich cookie as a standard unit, five moles of cookies would weigh roughly 0.5% of the Earth's mass or 20 times the mass of the ocean. Please consider checking the numbers used in this paragraph. You will likely want to use **scientific notation**. 

As you may have guessed, a mole is not a particularly useful number for counting cookies. However, it is very useful number for counting atoms and molecules. Why? Well, one reason is because atoms and molecules are very small. A second reason is that the mole helps us to convert **units of numbers** of atoms and molecules (i.e., moles) to **units of mass** (i.e., grams). Based on the current definintion of the mole, **one mole of carbon-12 atoms has a mass of 12 g** (or a dozen grams). Please note that, on average, **one mole of oxygen atoms has a mass of 15.999 g** and, on average, **one mole of hydrogen atoms has a mass of 1.008 g**.

## 3.2. The Composition of Atoms
**Atoms** are made up of positively charged **protons**, neutrally charged **neutrons**, and negatively charged **electrons**. The mass of a proton is 1.00727647 g/mol. The mass of a neutron is 1.008664 g/mol. The mass of an electron is 0.00054858 g/mol. The electrical charge associated with a proton is 1.6021766208 × 10<sup>−19</sup> Coulombs. The electrical charge associated with an electron is 1.6021766208 × 10<sup>−19</sup> Coulombs.  

* The **elemental identity** of an atom/ion is determined by the number of protrons.  
* The **charge** of an atom/ion is determined by the number of protons minus the number of electrons.
* The **mass** of an atom/ion in g/mol is almost exactly equal to the number of protons plus the number of neutrons.

For example, a carbon-12 atom contains 6 protons, 6 neutrons, and 6 electrons.
* The **elemental identity is carbon** because all atoms containing 6 protons are the element carbon and will exhibit nearly identical chemical behavior.
* The **charge is 0** because (6 protons × 1.6021766208 × 10<sup>−19</sup> Coulombs/proton) - (6 electrons × 1.6021766208 × 10<sup>−19</sup> Coulombs/electron) = 0
* The **mass is 12 g/mol** because (6 protons × 1.00727647 g/mol protons) + (6 neutrons × 1.008664 g/mol neutrons) = 12 g/mol carbon (?)

Actually, that last one is a little tricky. The amu is defined such that the mass of one carbon-12 atom *is* equal to 12 amu. However, this definition means that the masses of protons and neutrons must be *slightly greater* than 1 amu and that the mass of any atom that is not carbon-12 will only be *approximately equal* to the number of protons plus the number of neutrons.   

**Why?** It turns out that, as an atomic nucleus is created from separate protons and neutrons, energy is either released (in the case of small atoms, like helium), or absorbed (in the case of large atoms, like uranium). This energy doesn't come out of nowhere, but is instead converted from mass according to Einstein's famous equation, $ E  = mc^2 $, where $E$ is the energy equivalent of the mass gained or lost, $m$ is the mass and $c$ is the speed of light in a vacuum ($c = 3\times 10^8 m/s$).

## 3.3.  Converting from mass to number
The next step in our quest to develop a recipe for seawater is to convert the table of Major Seawater Constituents from **units of mass** to **units of moles** (i.e., numbers of ions). Why? Because the chemical formulas of the salts we want to use show the ratios of numbers of ions, not the ratios of masses.  

For example, one mole of calcium carbonate (CaCO<sub>3</sub>) contains:  
* 1 mole of calcium ion (Ca<sup>2+</sup>) 
* 1 mole of carbonate ion (CO<sub>3</sub><sup>2-</sup>)  

In turn, one mole of carbonate ion (CO<sub>3</sub><sup>2-</sup>) contains:  

* 1 mole of carbon (C) 
* 3 moles of oxygen (O).  

However, the molar mass of calcium is 40.078 g, the molar mass of carbon is 12.011 g, and the molar mass of oxygen is 15.999 g. Hence, one mole of carbonate contains:  

* 40.078 g of calcium
* 12.011 g of carbon
* 3 × 15.999 g = 47.997 g of oxygen

The molar mass of carbonate (CO<sub>3</sub><sup>2-</sup>) is $12.011g + 3(15.999g) = 60.008 g$  
The molar mass of calcim carbonate (CaCO<sub>3</sub><sup>2-</sup>) is $40.078g + 12.011g + 3(15.999g) = 100.086 g$

```python
#This is an example code block demonstrating a calculation of the molar mass of carbonate
#Please use similar code blocks in composing your answers
calcium_molar_mass_g = 40.078
carbon_molar_mass_g = 12.011
oxygen_molar_mass_g = 15.999
carbonate_molar_mass_g = carbon_molar_mass_g + 3*oxygen_molar_mass_g
print(carbonate_molar_mass_g)
calcium_carbonate_molar_mass_g = calcium_molar_mass_g + carbon_molar_mass_g + 3*oxygen_molar_mass_g
print(calcium_carbonate_molar_mass_g)
```

$$ Concentration\left(\frac{mol}{kg}\right) = \frac{Concentration\left(\frac{\color{red}{g}}{kg}\right)}{molar \space mass\left(\frac{\color{red}{g}}{mol}\right)}$$  
As you can see, the grams (in <font color="red"> **red** </font>) cancel out, leaving mol/kg.

## 5.1. From Salts to Ions
To calculate the composition of our proposed recipe in terms of costituent **ions**, we need to consider the ion contribution of each of our proposed ingredients. For example, if were to take **1 mol of sodium chloride (NaCl)** and add water up the a mass of 1 kg, and this salt were to fully dissolved, then we would get a **sodium (Na<sup>+</sup>)** concentration of 1 mol/kg and a **chloride (Cl<sup>-</sup>)** concentration of 1 mol/kg. 

However, if we were to start with **0.5 mol of sodium chloride (NaCl)** and **0.5 mol of calcium chloride (CaCl<sub>2</sub>)** and add water up the a mass of 1 kg, and these salts were to fully dissolve, then we would end up with a **sodium (Na<sup>+</sup>)** concentration of 0.5 mol/kg, a **calcium (Ca<sup>2+</sup>)** concentration of 0.5 m0l/kg and a **chloride (Cl<sup>-</sup>)** concentration of 1.5 mol/kg.  

To calculate the concentrations of these ions, we multiply the amount of salt added (in moles) by the number of chloride ions for each mole of salt and sum over all of the salts. For example:

$$Total \space Cl^- \left( \frac{mol}{kg}\right)  = 
\frac{1 \cdot NaCl \space (mol)  + 2 \cdot CaCl_2 \space (mol)}
     {Total \space Mass \space (kg)}$$  

In this case, the total **chloride (Cl<sup>-</sup>)** concentration is 1.5 mol/kg, which is **2.80 times** the actual **chloride (Cl<sup>-</sup>)** concentration of seawater, or a percent error of **180%**. 

Obviously, doing these calculations by hand or handheld calculator is possible. However, we will be wanting to calculate the concentration of  each ion over multiple salts (e.g., NaCl, CaCl<sub>2</sub>, KCl, MgCl<sub>2</sub>) for each proposed recipe and testing results for multiple proposed recipes. Overall, this process is:
* repetative
* tedious
* prone to error.    

Hence, we will be **using the computer** to do this for us.

## 5.2. Matrix Multiplication  
The specific technique we will be **using the computer** to perform is **matrix multiplication**. While it sounds scary, **matrix multiplication** is just a repetative form of **addition** and **multiplication**. For example, if we have a solution of 1 kg formulated from water and:
* 1 mol of Sodium Chloride (NaCl)
* 0.5 mol of Magnesium Chloride (MgCl<sub>2</sub>)
* 1 mol of Sodium Sulfate (Na<sub>2</sub>SO<sub>4</sub>) 
* 0.5 mol of Magnesium Sulfate (MgSO<sub>4</sub>)  

Then, we could set up a **series of equations** similar to the one above for chloride:  

$$Total \space Cl^- \left( \frac{mol}{kg}\right)  = 
\frac{1 \cdot NaCl \space (mol)  + 2 \cdot MgCl_2 \space (mol)}
     {Total \space Mass \space (kg)} = 
\frac{1 \cdot (1 \space mol)  + 2 \cdot (0.5 \space mol)}
     {1 \space kg} = 
\frac{2 \space mol}{kg}$$  
     
$$Total \space Na^+ \left( \frac{mol}{kg}\right)  = 
\frac{1 \cdot NaCl \space (mol)  + 2 \cdot Na_2SO_4 \space (mol)}
     {Total \space Mass \space (kg)} = 
\frac{1 \cdot (1 \space mol)  + 2 \cdot (1 \space mol)}
     {1 \space kg} = 
\frac{3 \space mol}{kg}$$  
     
$$Total \space Mg^{2+} \left( \frac{mol}{kg}\right)  = 
\frac{2 \cdot MgCl_2 \space (mol)  + 1 \cdot MgSO_4 \space (mol)}
     {Total \space Mass \space (kg)} = 
\frac{2 \cdot (0.5 \space mol)  + 1 \cdot (0.5 \space mol)}
     {1 \space kg} = 
\frac{1.5 \space mol}{kg}$$   
     
$$Total \space SO_4^{2-} \left( \frac{mol}{kg}\right)  = 
\frac{1 \cdot Na_2SO_4 \space (mol)  + 1 \cdot MgSO_4 \space (mol)}
     {Total \space Mass \space (kg)} = 
\frac{1 \cdot (1 \space mol)  + 1 \cdot (0.5 \space mol)}
     {1 \space kg} = 
\frac{1.5 \space mol}{kg}$$   

**Matrix multiplication** just gives us a simpler way of writing all of these equations. For example, we could write one matrix (or "vector" because it is only a single column) containing recipe of ingredients:  

$$recipe\_vector = \begin{pmatrix}
NaCl \space (mol) \\
MgCl_2 \space (mol) \\
Na_2SO_4 \space (mol) \\
MgSO_4 \space (mol) \\
\end{pmatrix} = \begin{pmatrix}
1   \\
0.5 \\
1   \\
0.5 \\
\end{pmatrix}$$  

We can then write another matrix containing the ion composition of each of these ingredients:  

$$salt\_matrix = \begin{pmatrix}
1 & 1 & 0 & 0 \\
2& 0 & 1 & 0 \\
0 & 2 & 0 & 1 \\
0 & 0 & 1 & 1 \\
\end{pmatrix}$$    
  
where the **rows** correspond to the **salts**:  
* Sodium Chloride(NaCl)
* Magnesium Chloride (MgCl<sub>2</sub>)
* Sodium Sulfate (Na<sub>2</sub>SO<sub>4</sub>) 
* Magnesium Sulfate (MgSO<sub>4</sub>)    

and the **columns** correspond to the **ions**:    

* chloride (Cl<sup>-</sup>) 
* sodium (Na<sup>+</sup>)
* magnesium (Mg<sup>2+</sup>)
* sulfate (SO<sub>4</sub><sup>2-</sup>)

Please **pause here** to make sure that you understand how the recipe and chemical formulas of the ingredients have been translated to matrix/vector form. 

### 5.2.1. Performing the Matrix Multiplication  
Now that we have the recipe vector:   
$$recipe\_vector = \begin{pmatrix}
1   \\
0.5 \\
1   \\
0.5 \\
\end{pmatrix}$$   
and the salt matrix:  

$$salt\_matrix = \begin{pmatrix}
1 & 1 & 0 & 0 \\
2& 0 & 1 & 0 \\
0 & 2 & 0 & 1 \\
0 & 0 & 1 & 1 \\
\end{pmatrix}$$    

We can **matrix multiply** the transposed recipe vector by the salt matrix to get the composition of the solution:

$$\begin{pmatrix}
1 & 0.5 & 1 & 0.5 \\
\end{pmatrix}
%
\begin{pmatrix}
1 & 1 & 0 & 0 \\
2& 0 & 1 & 0 \\
0 & 2 & 0 & 1 \\
0 & 0 & 1 & 1 \\
\end{pmatrix}
=
\begin{pmatrix}
1(1) + 0.5(2) + 1(0) + 0.5(0) \\
1(1) + 0.5(0) + 1(2) + 0.5(0) \\
1(0) + 0.5(2) + 1(0) + 0.5(1) \\
1(0) + 0.5(0) + 1(1) + 0.5(1) \\
\end{pmatrix}
= 
\begin{pmatrix}
2 \\
3 \\
1.5   \\
1.5 \\
\end{pmatrix}
$$
